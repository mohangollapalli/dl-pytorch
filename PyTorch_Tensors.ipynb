### Tensor - Numpy Interoperability

```python
# Create a Tensor from a numpy array
t = torch.from_numpy(np.arange(0,10,1))
t
```

```python
# Reshape the 1-d tensor to 2-d tensor using reshape/view method
t.reshape(2,5)
```

```python
t.view(2,5)
```

```python
# Convert the tensor object back to a numpy array
t.numpy()
```

```python
t.tolist()
```

### Basic Tensor Operations

- Tensor Addition
- Scalar Multiplication
- Tensor Multiplication (Hadamard Product)
- Matrix Multiplication

```python
u = torch.tensor([1, 2, 3])
v = torch.tensor([1, 1, 0])
u, v
```

```python
# Vector dot-product
torch.dot(u, v)
```

```python
# Scalar multiplication
u * 2
```

```python
# Scalar addition
u + 2
```

```python
# Element-wise addition of two vectors (1-d tensors)
u + v
```

```python
# Element-wise multiplication of two vectors (Hadamard Product)
u * v
```

```python
# Operations on 2-D Tensors
t1 = torch.tensor([[1, 2], [1, 1]])
t2 = torch.tensor([[1, 0], [0, 1]])
```

```python
t1 + t2
```

```python
# Hadamard Product
t1 * t2
```

```python
# Matrix multiplication using 2-d Tensors
torch.mm(t1, t2)
```
